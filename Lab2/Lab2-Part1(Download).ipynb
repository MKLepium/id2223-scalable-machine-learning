## Load Dataset

```python
from datasets import load_dataset, DatasetDict

common_voice = DatasetDict()

common_voice["train"] = load_dataset("mozilla-foundation/common_voice_11_0", "de", split="train+validation")
common_voice["test"] = load_dataset("mozilla-foundation/common_voice_11_0", "de", split="test")
#common_voice["train"] = load_dataset("openai/whisper-large-v3", "de", split="train+validation", use_auth_token=True)
#common_voice["test"] = load_dataset("openai/whisper-large-v3", "de", split="test", use_auth_token=True)


print(common_voice)
```

We get rid of meta information as well as general stuff that we don't need.

```python
common_voice = common_voice.remove_columns(["accent", "age", "client_id", "down_votes", "gender", "locale", "path", "segment", "up_votes"])

print(common_voice)
```
